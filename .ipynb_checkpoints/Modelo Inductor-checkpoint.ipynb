# Sobre el calentamiento de un cilindro hueco por inducción electromagnética

```python
%matplotlib inline
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg','pdf')

import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc

plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Times'
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['figure.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['lines.markersize'] = 3
```

```python
R = 1
RbyD = np.arange(2, 10, 1)
rByR = np.linspace(bByR, 1.0, 1024)
delta = R/RbyD

m = np.sqrt(2)/delta
mur = 1
```

```python
def computeXbyXs(m, R, bByR, rByR, mur):
   
    r = rByR*R
    b = bByR*R

    mR = m*R
    mr = m*r[:,np.newaxis]    
    
    b = bByR*R
    mb = m*b
    
    BbyA = -(2*mur*(sc.berp(mb) + 1j*sc.beip(mb)) - 1j*mb*(sc.ber(mb) + 1j*sc.bei(mb)))/ \
        (2*mur*(sc.kerp(mb) + 1j*sc.keip(mb)) - 1j*mb*(sc.ker(mb) + 1j*sc.kei(mb)))
  
    Ch = (1 + BbyA*(sc.ker(mr) + 1j*sc.kei(mr))/(sc.ber(mr) + 1j*sc.bei(mr)))/ \
            (1 + BbyA*(sc.ker(mR) + 1j*sc.kei(mR))/(sc.ber(mR) + 1j*sc.bei(mR)))        
    HbyHs = (sc.ber(mr) + 1j*sc.bei(mr))/(sc.ber(mR) + 1j*sc.bei(mR))*Ch
    
    Cj = (1 + BbyA*(sc.kerp(mr) + 1j*sc.keip(mr))/(sc.berp(mr) + 1j*sc.beip(mr)))/ \
            (1 + BbyA*(sc.kerp(mR) + 1j*sc.keip(mR))/(sc.berp(mR) + 1j*sc.beip(mR)))
    JbyJs = (sc.berp(mr) + 1j*sc.beip(mr))/(sc.berp(mR) + 1j*sc.beip(mR))*Cj
    
    return np.abs(HbyHs), np.abs(JbyJs), rByR
    
```

```python
bByR = 0.6
HbyHs, JbyJs, rByR = computeXbyXs(m, R, bByR, rByR, mur)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 3.5))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, 
                    top=0.9, wspace=0.3, hspace=0.4)

lines = ax1.plot(rByR, JbyJs)
ax1.invert_xaxis()
ax1.set_xlabel(r'$r/R$')
ax1.set_ylabel(r'$|J/J_s|$')
plt.xlim([1, rByR[0]])

for i in range(len(RbyD)):
    text[i] = r'$R/{\delta} = $ ' + str(RbyD[i])
plt.legend(lines, text, bbox_to_anchor=(1.1, 0.75))

ax2.plot(rByR, HbyHs)
ax2.invert_xaxis()
ax2.set_xlabel(r'$r/R$')
ax2.set_ylabel(r'$|H/H_s|$')
plt.xlim([1, rByR[0]])

fig.suptitle(r'$b/R = $ ' + str(bByR))
plt.show()
```

```python
dByDelta = np.linspace(1, 15, 1024)
mur = np.linspace(1, 1000, 1024)
d = 1
```

```python
def computePower(dByDelta, tByD, mur, d):    
        
    t = tByD*d
    b = 0.5*t
    delta = d/dByDelta
    m = np.sqrt(2)/delta
    mb = m*b
    R = d/2
    
    BbyA = -(2*mur*(sc.berp(mb) + 1j*sc.beip(mb)) - 1j*mb*(sc.ber(mb) + 1j*sc.bei(mb)))/ \
        (2*mur*(sc.kerp(mb) + 1j*sc.keip(mb)) - 1j*mb*(sc.ker(mb) + 1j*sc.kei(mb)))
      
    Cp = (1 + BbyA*(sc.kerp(m*R) + 1j*sc.keip(m*R))/(sc.berp(m*R) + 1j*sc.beip(m*R)))/ \
            (1 + BbyA*(sc.ker(m*R) + 1j*sc.kei(m*R))/(sc.ber(m*R) + 1j*sc.bei(m*R)))
    
    P = 2/(m*R)*(sc.berp(m*R) + 1j*sc.beip(m*R))/(sc.ber(m*R) + 1j*sc.bei(m*R))*Cp
    
    pt = P.real
    qt = P.imag
    
    return pt, qt
    
```

```python
t = tByD*d
b = 0.5*t
delta = d/dByDelta
m = np.sqrt(2)/delta
mb = m*b
R = d/2
    
BbyA = -(2*mur*(sc.berp(mb) + 1j*sc.beip(mb)) - 1j*mb*(sc.ber(mb) + 1j*sc.bei(mb)))/ \
        (2*mur*(sc.kerp(mb) + 1j*sc.keip(mb)) - 1j*mb*(sc.ker(mb) + 1j*sc.kei(mb)))
      
```

```python
pt, qt = computePower(dByDelta, tByD, mur, d)
plt.plot(dByDelta, pt.T)
```
